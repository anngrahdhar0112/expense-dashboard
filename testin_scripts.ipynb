```python
import pandas as pd

# 1. Read the CSV
file_upload = pd.read_csv('data/clean/superstore_sales.csv', encoding='latin1')

# 2. Helper function to fix a mixed-format date column
def fix_mixed_dates(df, colname):
    s = df[colname].astype(str).str.strip()

    # Rows like 2016-08-11  → YYYY-MM-DD
    mask_dash = s.str.contains('-')

    # Rows like 11/11/2016, 1/21/2014, 3/03/2017 → MM/DD/YYYY
    mask_slash = s.str.contains('/')

    d_dash = pd.to_datetime(s.where(mask_dash),
                            format='%Y-%m-%d',
                            errors='coerce')
    d_slash = pd.to_datetime(s.where(mask_slash),
                             format='%m/%d/%Y',
                             errors='coerce')

    # Use dash-parsed where available, otherwise slash-parsed
    df[colname] = d_dash.combine_first(d_slash).dt.date

# 3. Fix both date columns
fix_mixed_dates(file_upload, "Order Date")
fix_mixed_dates(file_upload, "Ship Date")

# (Optional) quick sanity check for any remaining bad dates
print("Order Date nulls:", file_upload["Order Date"].isna().sum())
print("Ship Date nulls:", file_upload["Ship Date"].isna().sum())
```
